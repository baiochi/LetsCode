## dados

```python
import pandas as pd
```

```python
df_titanic = pd.read_csv('./dados/titanic.csv')
df_titanic.head(2)
```

## barplot

```python
sexo_series = df_titanic.groupby(['Sex']).size().copy()
sexo_series
```

```python
plt.bar(sexo_series.index, sexo_series);
```

```python
#22f5f5 #30ff6e
plt.barh(sexo_series.index, sexo_series, color=['#8c8c8c', '#4c5dfc']);
```

Mais homens sobreviveram no titanic do que mulheres?

```python
sexo_sobreviveu_series = df_titanic.groupby(by=['Sex'])['Survived'].sum()
sexo_sobreviveu_series
```

```python
#22f5f5 #30ff6e

plt.barh(sexo_series.index, sexo_series, color=['#8c8c8c'], label='# pessoas');
plt.barh(sexo_sobreviveu_series.index, 
         sexo_sobreviveu_series, 
         color=['#4c5dfc'], 
         label='# pessoas sobreviventes');
plt.legend();
```

## scatter

```python
plt.scatter(x=df_titanic['Age'], y=df_titanic['Fare']);

plt.xlabel('Idade');
plt.ylabel('Preo da passagem');
```

## pie (pizza)

```python
df_titanic['Sex'].value_counts()
```

```python
plt.pie(df_titanic['Sex'].value_counts(), 
        labels=df_titanic['Sex'].value_counts().index, 
        autopct='%1.1f%%', 
        shadow=True, startangle=90);
```

## histograma

```python
plt.hist(x=df_titanic['Age']);

plt.title('histograma da idade');

plt.xlabel('Idade');
plt.ylabel('quantidade');
```

```python
plt.hist(x=df_titanic['Age'], 
         color='blue', label='Total');
plt.hist(x=df_titanic[df_titanic['Survived']==True]['Age'], 
         color='red', label='Sobreviveram');
         
plt.title('histograma da idade');
plt.legend();
plt.xlabel('Idade');
plt.ylabel('quantidade');
```

## boxplot

```python
plt.boxplot(x=df_titanic['Age'].dropna());
plt.ylabel('Idade');
```
