## 1. Pilha



- É um conjunto dinâmico
- LIFO: *Last In, First Out* $\rightarrow$ o que entra por último é o primeiro a sair

### Operações básicas

- pop(): Remove o elemento do topo da pilha (primeiro elemento)
- push(item): Adiciona um item no topo da pilha
- peek(): Retorna o topo da pilha
- is_empty(): Verifica se a pilha está vazia

### Exemplos de uso:

- Função desfazer
- Backtracking (para avaliar a execução de um código, por exemplo)

#### E como fazer em python? Vamos ver um jeito simples de criar uma pilha em python usando a estrutura collections.deque

Vale dizer que também poderiamos usar uma lista comum.

```python
from collections import deque

stack = deque()

## Push
stack.append(3)
stack.append(4)
stack.append(2)
print(stack)

## Pop
print(stack.pop())
print(stack)

stack.append(6)
print(stack)

print(stack.pop())
print(stack)

print(stack.pop())
print(stack.pop())
stack.pop()
```

## 2. Fila



- É um conjunto dinâmico
- FIFO: *First In, First Out* $\rightarrow$ o que entra primeiro é o primeiro a sair

### Operações básicas

- enqueue(k): Insere o elemento k na última posição da fila
- dequeue(): Remove o primeiro elemento da fila
- is_empty(): Verifica de a fila está vazia

### Exemplos de uso:

- Fila de processos de um computador
- Sistema que realiza o processamento de pedidos e necessita que eles saiam de acordo com a ordem que entraram

#### E como fazer em python? Vamos ver um jeito simples de criar uma pilha em python usando a estrutura collections.deque

```python
from collections import deque

queue = deque()

## Enqueue
queue.append(3)
queue.append(4)
queue.append(2)
print(queue)

## Dequeue
print(queue.popleft())
print(queue)

queue.append(6)
print(queue)

print(queue.popleft())
print(queue)

print(queue.popleft())
print(queue.popleft())
# queue.pop()
```
