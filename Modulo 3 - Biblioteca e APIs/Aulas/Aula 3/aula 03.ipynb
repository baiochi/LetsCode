### ```pd.DataFrame.drop()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html  
  
   
Feito para dropar colunas e/ou linhas de um DataFrame.

````python
# dropa a linha 0
df_titanic.drop(0)
````

É interessante ver que a função não altera o DataFrame, mas sim, retorna o valor.  
  
Para alterarmos o DataFrame, será necessário atribuir o resultado a uma variável?  
É uma opção. Mas também, existe um parâmetro que chama ```ìnplace```, que se for ```True```, muda o DataFrame original.

````python
# dropa a linha cujo o index==0
df_titanic.drop(0, inplace=True)
````

````python
# não existe mais o index=0
df_titanic.head()
````

````python
# dropa as linhas cujo o index == 1 e index == 2
df_titanic.drop([1, 2], inplace=True)
df_titanic.head()
````

````python
# dropa a coluna 'Cabin', lembrando que o axis tem que ser 1
df_titanic.drop('Cabin', axis=1, inplace=True)
df_titanic.head()
````

````python
# dropa a coluna 'SibSp' e 'Parch', lembrando que o axis tem que ser 1
df_titanic.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df_titanic.head()
````

````python
# dropa a coluna 'Cabin', lembrando que o argumento aqui usado é "columns" e tem que ser especificado
df_titanic.drop(columns=['Ticket', 'Embarked'], inplace=True)
df_titanic.head()
````

### ```pd.DataFrame.rename()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html  
  
   
Renomear colunas ou indices.

````python
# renomeia as colunas e retorna um df
# no dict não precisa ter todas as colunas, só as de interesse
df_titanic.rename(columns={
    'PassengerId': 'IdPassageiro',
    'Survived': 'Sobreviveu',
    'Name': 'Nome',
    'Sex': 'Sexo',
    'Age': 'Idade',
    'Fare': 'Tarifa'
})
````

````python
# para sobrescrever o df original, precisa usar o inplace=True
df_titanic.rename(columns={
    'PassengerId': 'IdPassageiro',
    'Survived': 'Sobreviveu',
    'Name': 'Nome',
    'Sex': 'Sexo',
    'Age': 'Idade',
    'Fare': 'Tarifa'
}, inplace=True)
df_titanic.head()
````

### ```pd.DataFrame.sort_values()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html  
  
   
Ordena o DataFrame

````python
# ordena as colunas de forma crescente e retorna um df
# para sobrescrever, precisa do inplace=True
df_titanic.sort_values(by=['Idade'])
````

````python
# ordena as colunas de forma decrescente e retorna um df 
# para sobrescrever, precisa do inplace=True
df_titanic.sort_values(by=['Idade'], ascending=False)
````

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_index.html  
  
   
Ordena o DataFrame pelo index

````python
# ordena o df pelo indice de forma decrescente e retorna um df
# para sobrescrever, precisa do inplace=True
df_titanic.sort_index(ascending=False)
````

````python
# ordena o df pelo indice de forma crescente. 
# para sobrescrever, precisa do inplace=True
df_titanic.sort_index()
````

---
### ```pd.DataFrame.reset_index()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html  
  
   
Reseta os indices

````python
# aqui temos a idade como index, mas queremos usá-la como coluna
df_titanic.groupby(by=['Idade'])['Sobreviveu'].sum()
````

````python
# como fazer?
# com a função reset_index()
df_titanic.groupby(by=['Idade'])['Sobreviveu'].sum().reset_index()
````

````python
# mas e se eu não quisesse guardar o indice antigo?
# uso o parâmetro drop=True
df_titanic.groupby(by=['Idade'])['Sobreviveu'].sum().reset_index(drop=True)
````

### Funções úteis e gerais:
* ```pd.DataFrame.max()```: valor máximo de uma determinada Series ou DataFrame (traz de todas as Series do df);
* ```pd.DataFrame.min()```: valor mínimo de uma determinada Series ou DataFrame (traz de todas as Series do df);
* ```pd.DataFrame.mean()```: média de uma determinada Series ou DataFrame (traz de todas as Series do df);
* ```pd.DataFrame.median()```: mediana de uma determinada Series ou DataFrame (traz de todas as Series do df);
* ```pd.DataFrame.sum()```: somatório de uma determinada Series ou DataFrame (traz de todas as Series do df);
* ```pd.DataFrame.cumsum()```: somatório acumulado de uma determinada Series ou DataFrame (traz de todas as Series do df);
* ```pd.DataFrame.quantile()```: quantil de uma determinada Series ou DataFrame (traz de todas as Series do df);
* ```pd.DataFrame.describe()```: estatísticas gerais de uma determinada Series ou DataFrame (traz de todas as Series do df).

## Limpeza de dados

### ```pd.DataFrame.isna()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isna.html  
  
  
Verifica se a series de um df é nulo (NaN) ou não

````python
df_titanic.isna()
````

Como otimizar isso?

O df resultante é de booleanos, com isso, podemos somá-lo e chegar no resultado da soma pela coluna!

````python
df_titanic.isna().sum()
````

E se dermos um ```reset_index()```, ele se transforma em um df.

````python
df_titanic.isna().sum().reset_index()
````

Podemos fazer isso pra Series também:

````python
df_titanic['Idade'].isna()
````

````python
df_titanic['Idade'].isna().sum()
````

### ```pd.DataFrame.notna()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.notna.html  
  
  
Detecta as células que não são nulas

````python
df_titanic.notna()
````

Semelhantemente ao comando anterior:

````python
df_titanic.notna().sum()
````

E se dermos um ```reset_index()```, ele se transforma em um df.

````python
df_titanic.notna().sum().reset_index()
````

Podemos fazer a mesma coisa pra Series também:

````python
df_titanic['Idade'].notna().sum()
````

### ```pd.DataFrame.fillna()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html  
  
  
Preenche os nulos (NaN) com o valor desejado

````python
# vai preencher TODOS os nulos do df com 0
df_titanic.fillna(0)
````

````python
# vai preencher TODOS os nulos do df utilizando os valores mapeados no dicionário "values"
valores = {"Idade": 18, "Tarifa": 50}
df_titanic.fillna(valores)
````

Lembrando que o resultado é um df, para sobrescrever o DataFrame original, precisa setar o argumento ```inplace=True```.

Uma aplicação muito legal, é que podemos dar o ```fillna``` direto na Series:

````python
# preenchendo a Idade com a média
df_titanic['Idade'].fillna(df_titanic['Idade'].mean())
````

### ```pd.DataFrame.dropna()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html  
  
  
Preenche os nulos (NaN) com o valor desejado

````python
# dropa as linhas que tem pelo menos 1 nulo "missing"
df_titanic.dropna()
````

````python
# dropa as colunas que tem pelo menos 1 nulo "missing"
df_titanic.dropna(axis='columns')
````

````python
# dropa as linhas onde todos os elementos são nulos
df_titanic.dropna(how='all')
````

````python
# dropa as linhas que tem mais de 2 nulos "missing"
df_titanic.dropna(thresh=2)
````

````python
# a mesma coisa em relação às funções anterior, 
# mas ao invés de olhar o df inteiro, olha apenas as colunas especificadas
df_titanic.dropna(subset=['Idade', 'Tarifa'])
````

### ```pd.Series.map()```  
https://pandas.pydata.org/docs/reference/api/pandas.Series.map.html  
  
  
Mapeia valores para outros

````python
# usa o dicionário para mapear
# retorna uma Series nova e para sobrescrever, precisa atribuir de novo (não existe o parâmetro inplace)
mapa = {
    'female': 'feminino',
    'male': 'masculino'
}
df_titanic['Sexo'].map(mapa)
````

### ```pd.DataFrame.duplicated()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.duplicated.html  
  
  
Retorna um df de booleanos com linhas repetidas.
  
Por padrão, a primeira é setada como False e o restante como False.

````python
df_titanic.duplicated()
````

````python
# para ter a quantidade de linhas repetidades
df_titanic.duplicated().sum()
````

````python
# podemos setar um subconjunto de colunas usando o argumento "subset"
df_titanic.duplicated(subset=['Idade', 'Sexo'])
````

````python
# como escolhemos algo que não é "único", temos muitas linhas repetidas
df_titanic.duplicated(subset=['Idade', 'Sexo']).sum()
````

````python
# Usando o keep='last' a última ocorrência é setada como False e o restante como True
df_titanic.duplicated(keep='last')
````

### ```pd.DataFrame.drop_duplicates()```  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html  
  
  
Retorna um DataFrame com as linhas duplicadas removidas  
  
O padrão é manter a primeira ocorrência e remover as outras.

````python
# retorna um df com as linhas duplicadas (em relação a todas as colunas)
df_titanic.drop_duplicates()
````

````python
# mas conseguimos também, dropar as linhas em relação a algumas colunas apenas
df_titanic.drop_duplicates(subset=['Sexo', 'Idade'])
````

````python
# podemos manter apenas a útlima ocorrência ao invés da primeira!
df_titanic.drop_duplicates(subset=['Sexo', 'Idade'], keep='last')
````
