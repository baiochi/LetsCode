```python
import warnings; warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
```

Make all functions below without using `cv2` library

## Loading images

```python
image_1 = plt.imread('data/ednaldo.jpg')
image_2 = plt.imread('data/batman.jpeg')
image_3 = plt.imread('data/calculista.jpeg')
images = (image_1, image_2, image_3)
```

# Rotate image

Using `np.rot90` function

```python
fig, ax = plt.subplots(3,4, figsize=(20,12))
for axis, image in enumerate(images):
    for rotation in range(4):
        ax[axis, rotation].imshow(np.rot90(image, rotation));
```

# Change image to gray scale

```python
rgb_to_gray_scale = lambda x : np.dot(x[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
fig, ax = plt.subplots(1,3, figsize=(20,12))
for axis, image in enumerate(images):
    ax[axis].imshow(rgb_to_gray_scale(image), cmap='gray');
```

# Threshold images

```python
# Highlight high intensity colors
def threshold_image(image, threshold): 

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            pixel = image[i,j,:]
            if pixel.sum() < threshold:
                image[i,j,:] = [0] * 3
                
    return image
```

```python
fig, ax = plt.subplots(1,3, figsize=(20,12))
for axis, image in enumerate(images):
    ax[axis].imshow(threshold_image(image, 200));
```
