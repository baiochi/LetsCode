```python
import warnings; warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
```

## Loading images

```python
image_1 = plt.imread('data/ednaldo.jpg')
image_2 = plt.imread('data/batman.jpeg')
image_3 = plt.imread('data/calculista.jpeg')
images = (image_1, image_2, image_3)
```

# Show images

```python
fig, ax = plt.subplots(1,3, figsize=(16,12))
print('Ednaldo Pereira Foto Perfil')
for axis, image in enumerate(images):
    ax[axis].imshow(image);
```

# Negative Image

```python
fig, ax = plt.subplots(1,3, figsize=(16,12))
print('Ednaldo Pereira Multiverso')
for axis, image in enumerate(images):
    ax[axis].imshow(np.invert(image));
```

# Transpose Image matrix

```python
fig, ax = plt.subplots(1,3, figsize=(16,12))
print('Ednaldo Pereira Torto')
for axis, image in enumerate(images):
    transpose_image = np.transpose(image)
    transpose_image = np.swapaxes(transpose_image, 0, 1)
    transpose_image = np.swapaxes(transpose_image, 1, 2)
    ax[axis].imshow(transpose_image);
```

# Unique colormap

```python
def red_scale(image, saturation=0):
    red_image = image.copy()
    red_image[:, :, [1,2]] = saturation
    return red_image

def green_scale(image, saturation=0):
    green_image = image.copy()
    green_image[:, :, [0,2]] = saturation
    return green_image

def blue_scale(image, saturation=0):
    blue_image = image.copy()
    blue_image[:, :, [0,1]] = saturation
    return blue_image
```

```python
fig, ax = plt.subplots(3,3, figsize=(20,12))
print('Ednaldo Pereira Colorido')
for axis, image in enumerate(images):
    ax[axis, 0].imshow(red_scale(image))
    ax[axis, 1].imshow(green_scale(image))
    ax[axis, 2].imshow(blue_scale(image));
```

# Changing colors

```python
# Set image to gray scale
rgb_to_gray_scale = lambda x : np.dot(x[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
# Ednaldo favorite filter
def insta_filter(image):
    filter_image = image.copy()
    
    filter_image[filter_image < 50] += 50
    
    filter_image[filter_image > 230] = 230
    
    return filter_image
```

```python
fig, ax = plt.subplots(3,3, figsize=(20,12))
print('Ednaldo Pereira Filtro Instagram')
for axis, image in enumerate(images):
    ax[axis, 0].imshow(rgb_to_gray_scale(image), cmap='gray')
    ax[axis, 1].imshow(insta_filter(image))
    ax[axis, 2].imshow(red_scale(image, 125));
```
