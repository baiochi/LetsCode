```python
import librosa
import speech_recognition as sr
from os import path
from IPython.display import Audio
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Function to transcribe audio with Google Speech Recognition
def transcribe_audio(audio_file, language='pt-BR'):
    
    # Select audio to recognize
    r = sr.Recognizer()
    # Load audio file
    with sr.AudioFile(audio_file) as source:
        audio = r.record(source)

    # Calling Google Speech Recognition
    try:
        text = r.recognize_google(audio, language=language)
        print('Google Speech Recognition: ' + text)
    except sr.UnknownValueError:
        print('Google Speech Recognition didn\'t understand')
        text = ''
    except sr.RequestError as e:
        print('Error while calling Google Speech Recognition; {0}'.format(e))
        text = ''
    return text
```
