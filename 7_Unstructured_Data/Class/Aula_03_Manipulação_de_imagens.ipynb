```python
import numpy as np
import PIL.Image as pil

import matplotlib.pyplot as plt
```

```python
imagem = plt.imread('img.jpg', format='jpeg')
```

```python
rgba_image = pil.open('img.jpg')
rgb_image = rgba_image.convert('RGB')
rgb_image
```

```python
plt.figure(figsize = (20,6))

plt.imshow(imagem[:,:,:3])

plt.box(False)

plt.title("Imagem no python");
```

```python
img_inv = np.invert(imagem)
```

```python
plt.figure(figsize = (20,6))

plt.imshow(img_inv)

plt.box(False)

plt.title("Imagem inversa no python")
```

```python
img_transposta = np.transpose(imagem)

img_transposta = np.swapaxes(img_transposta, 0, 1)
img_transposta = np.swapaxes(img_transposta, 1, 2)

plt.figure(figsize = (6, 20))

plt.imshow(img_transposta)

plt.box(False)

plt.title("Imagem transposta no python")
```

```python
R = imagem.copy()

R[:, :, [1,2]] = 0

G = imagem.copy()

G[:, :, [0,2]] = 0

B = imagem.copy()

B[:, :, [0,1]] = 0

C = imagem.copy()

C[:, :, [0,1,2]] = C[:, :, [0,1,2]] * [0.299, 0.587, 0.114]
```

```python
plt.figure(figsize = (20, 6))

plt.subplot(1,4,1)
plt.title("Camada R")
plt.imshow(R)
plt.box(False)

plt.subplot(1,4,2)
plt.title("Camada G")
plt.imshow(G)
plt.box(False)

plt.subplot(1,4,3)
plt.title("Camada B")
plt.imshow(B)
plt.box(False)

plt.subplot(1,4,4)
plt.title("Camada C")
plt.imshow(C, cmap ='gray')
plt.box(False)
```
